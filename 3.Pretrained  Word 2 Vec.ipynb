```python
path=r"C:\Users\Admin\GoogleNews-vectors-negative300.bin"
path
# 300 is dimention varient means it considering 300 properties of every word and calculate vector of every word and 
# 300 no of properties it considering so that vector also that much size
```

```python
from gensim.models import KeyedVectors
model=KeyedVectors.load_word2vec_format(path,binary=True) # we have loaded one pretrained model in model variable
model
```

```python
model['bad'] # with the help of 300 properties of bad word these are the vectors of bad word
```

```python
model.similarity('bad','good') # we have seen cosine similarities between two word and its near to +1 bc context are same
```

```python
model.most_similar('king') # here it shows king related contexts and show cosine value of every word related with king
```

```python
model.doesnt_match(['bad','awful','terrible','tower']) # if we have many words and if there is one word whose context is different
                                                       #  from other words then we can remove that word in short we can remove odd man out
```
